```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from libs.utils import get_group_figure_child, get_group_figure_adult

plt.rc('font', family='calibri')

import warnings
warnings.filterwarnings("ignore")
```

```python
df_child = pd.read_excel('./data/mapping_dataset.xlsx', sheet_name='children_dataset')
df_child_id = pd.read_excel('./data/mapping_dataset.xlsx', sheet_name='Child_ID')

df_adult = pd.read_excel('./data/mapping_dataset.xlsx', sheet_name='adult_total_dataset')
df_adult_id = pd.read_excel('./data/mapping_dataset.xlsx', sheet_name='Final_mapping_Final')
```

```python
# This is the task of writing the title of the graph and drawing the graph by group.
key_groups = list(df_child_id.ID_group.unique())

# RT_metabolites : ID Name -> dict_id

_features = df_child_id.Features
_id = df_child_id.ID

dict_id_child = {_features[i] : _id[i] for i in range(len(_features))}

# ID_group : features -> dict_id_group

dict_group_child = {}
for g in key_groups:
    dict_group_child[g] = list(df_child_id[df_child_id.ID_group == g].Features)


key_groups = list(df_adult_id.ID_group.unique())

# RT_metabolites : ID Name -> dict_id

_features = df_adult_id.Features
_id = df_adult_id.ID

dict_id_adult = {_features[i] : _id[i] for i in range(len(_features))}

# ID_group : features -> dict_id_group

dict_group_adult = {}
for g in key_groups:
    dict_group_adult[g] = list(df_adult_id[df_adult_id.ID_group == g].Features)
```

```python
feature_order_PUFA = ['7.52_279.2328', '7.93_303.2325', '9.2_303.2325','9.05_327.2329']
```

```python
get_group_figure_child(df_child, dict_id_child, order=['C', 'U', 'P', 'D'], palette=['Green', 'Pink', 'Orange', 'Red'],
                         is_save_file=True, save_name='children_PUFA', ha='right', x_label_rotation='42', figsize=(15, 13), feature_order=feature_order_PUFA)
```
